#Q2 In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv.

```python
#fetching the pages url details:
page_url2=[]

for i in driver.find_elements_by_xpath('//li[@class="a-normal"]//a'):
    page_url2.append(i.get_attribute('href'))
    
#fetching the product details:

#prd_avalability2=[]
prd_url2=[]

#taking only 3 pages:
for j in page_url2[:3]:
    driver.get(j)
    time.sleep(3)
    
    #fetching the url to get all the details 
    for q in driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]'):
        prd_url2.append(q.get_attribute('href'))


time.sleep(3)

#fetching the product details:
prd_name2=[]
brand_name2=[]
prd_rating2=[]
prd_noofrating1=[]
prd_Price2=[]
exp_del2=[]
prd_return2=[]
otr_details2=[]
ord_availility2=[]


for l in prd_url2:
    driver.get(l)
   # time.sleep(3)
    
    #fetching the title
    try:
        prdname1 = driver.find_element_by_id('productTitle')
        prd_name2.append(prdname1.text)
    except NoSuchElementException:
        prd_name2.append('_')
        
     #fetching the Brand name    
    try:
        brandname2 = driver.find_element_by_xpath('//td[@class="a-size-base prodDetAttrValue"]')
        brand_name2.append(brandname2.text)
    except NoSuchElementException:
        brand_name2.append('_')
      
    #fetching the product Rating
    try:
        prdrating1 = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
        prd_rating2.append(prdrating1.text)
    except NoSuchElementException:
        prd_rating2.append('_')
    
    #fetching the No of rating
    try:
        noofrating1 = driver.find_element_by_xpath('//div[@class="a-row a-spacing-medium averageStarRatingNumerical"]//span')
        prd_noofrating1.append(noofrating1.text)
    except NoSuchElementException:
        prd_noofrating1.append('_')    
   
    #fetching the order availibility
    try:
        availility2 = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        ord_availility2.append(availility2.text)
    except NoSuchElementException:
        ord_availility2.append('_') 

    #fetching the product price    
    try:
        prdprice1 = driver.find_element_by_id('priceblock_ourprice')
        prd_Price2.append(prdprice1.text)
    except NoSuchElementException:
        prd_Price2.append('_')
    
    #fetching the expected delivery
    try:
        expdel2 = driver.find_element_by_id('ddmDeliveryMessage')
        exp_del2.append(expdel2.text)
    except NoSuchElementException:
        exp_del2.append('_')
    
    #fetching the product return date
    try:
        prdreturn2 = driver.find_element_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')
        prd_return2.append(prdreturn2.text)
    except NoSuchElementException:
        prd_return2.append('_') 
        
    #fetching the product description    
    try:
        otrdetails2 = driver.find_element_by_id('productDescription')
        otr_details2.append(otrdetails2.text.replace('\n',''))
    except NoSuchElementException:
        otr_details2.append('_')

time.sleep(3)
#creating the dataframe
amazon_1=pd.DataFrame({})
amazon_1['Product Name']=prd_name2
amazon_1['Brand Name']=brand_name2
amazon_1['Product Rating']=prd_rating2
amazon_1['No of Ratings']=prd_noofrating1
amazon_1['Product avalibility']=ord_availility2
amazon_1['Product Price']=prd_Price2
amazon_1['Excepted Delivery']=exp_del2
amazon_1['Product Return']=prd_return2
amazon_1['Full Description']=otr_details2
amazon_1['Product URL']=prd_url2


#making data at the center
#amazon_1=amazon_1.style.set_properties(**{'text-align': 'center'}).set_table_styles([ dict(selector='th', props=[('text-align', 'center')])])
amazon_1    


#exporting data to csv
amazon_1.to_csv('amazon_Scrapping.csv', index=False)
```

```python
#importing the driver from the crome
driver7=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the webpage url:
url=('https://www.digit.in/')
driver7.get(url)

time.sleep(3)

#clicking on laptops
enter7=driver7.find_element_by_xpath('/html/body/div[1]/div[2]/div[4]/ul/li[3]/a')
enter7.click()
time.sleep(5)

#clicking on best gaming laptop
enter_lap7=driver7.find_element_by_xpath('/html/body/div[6]/div/div[2]/div[2]/ul/li[10]/a')
enter_lap7.click()
time.sleep(3)


#Fatechin the Data from the webpage

Laptopname7=[]
OS7=[]
Display7=[]
Processor7=[]
Memory7=[]
weight7=[]
dimension7=[]
graphic7=[]
price7=[]
seller7=[]
discription7=[]

#Fetching the laptop name
try:
    for i in driver7.find_elements_by_xpath('//div[@class="Top10-Seller"]/table[1]/tbody/tr/td[1]'):
        Laptopname7.append(i.text)
except NoSuchElementException:
    Laptopname7.append('_') 


#Fetching the OS
try:
    for t in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[3]/td[3]'):
        OS7.append(t.text)
except NoSuchElementException:
    OS7.append('_')  


#Fetching the display details    
try:
    for a in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[4]/td[3]'):
        Display7.append(a.text)
except NoSuchElementException:
    Display7.append('_')  

#Fetching the processor details    
try:
    for b in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[5]/td[3]'):
        Processor7.append(b.text)
except NoSuchElementException:
    Processor7.append('_')  

    #Fetching the memory details
try:
    for c in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[6]/td[3]'):
        Memory7.append(c.text)
except NoSuchElementException:
    Memory7.append('_')

#Fetching the weight details
try:
    for d in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[7]/td[3]'):
        weight7.append(d.text)
except NoSuchElementException:
    weight7.append('_')  

#Fetching the dimension details    
try:
    for e in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[8]/td[3]'):
        dimension7.append(e.text)
except NoSuchElementException:
    dimension7.append('_')  
    
#Fetching the grapic details    
try:
    for f in driver7.find_elements_by_xpath('//div[@class="Spcs-details"]/table[1]/tbody/tr[9]/td[3]'):
        graphic7.append(f.text)
except NoSuchElementException:
    graphic7.append('_')  

#Fetching the price details
try:
    for v in driver7.find_elements_by_xpath('//div[@class="Top10-Seller"]/table[1]/tbody/tr/td[3]'):
        price7.append(v.text)
except NoSuchElementException:
    price7.append('_')    

#Fetching the seller details
try:
    for p in driver7.find_elements_by_xpath('//div[@class="Top10-Seller"]/table[1]/tbody/tr/td[2]'):
        seller7.append(p.text)
except NoSuchElementException:
    seller7.append('_') 

#Fetching the discription details
try:
    for i in driver7.find_elements_by_xpath('//div[@class="Section-center"]//p'):
        discription7.append(i.text)
except NoSuchElementException:
    discription7.append('_') 
    
    
time.sleep(3)
#creating the dataframe
digit7=pd.DataFrame({})
digit7['Laptop Name']=Laptopname7
digit7['OS']=OS7
digit7['Display']=Display7
digit7['Processor']=Processor7 
digit7['Memory']=Memory7
digit7['Weight']=weight7
digit7['Dimension']=dimension7
digit7['Graphic']=graphic7
digit7['Price']=price7
digit7['Seller']=seller7
digit7['Discription']=discription7[:7] 


#making data at the center
#amazon_1=amazon_1.style.set_properties(**{'text-align': 'center'}).set_table_styles([ dict(selector='th', props=[('text-align', 'center')])])     
#display data frame
digit7
```

#8 Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#importing the driver from the crome
driver8=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the webpage url:
url=('https://www.forbes.com/billionaires/')
driver8.get(url)

time.sleep(5)

#fetching the data from the pages
Name8=[]
Rank8=[]
networth8=[]
citizion8=[]
age8=[]
source8=[]
industry8=[]

#Fetching the Name details
try:
    for i in driver8.find_elements_by_xpath('//div[@class="personName"]'):
        Name8.append(i.text)
except NoSuchElementException:
    Name8.append('_')

#Fetching the Rank details    
try:
    for i in driver8.find_elements_by_xpath('//div[@class="rank"]'):
        Rank8.append(i.text.replace('.',''))
except NoSuchElementException:
    Rank8.append('_')  

#Fetching the Networth details    
try:
    for k in driver8.find_elements_by_xpath('//div[@class="netWorth"]'):
        networth8.append(k.text)
except NoSuchElementException:
    networth8.append('_')    
#class="countryOfCitizenship"   

#Fetching the Citizon details
try:
    for m in driver8.find_elements_by_xpath('//div[@class="countryOfCitizenship"]'):
        citizion8.append(m.text)
except NoSuchElementException:
    citizion8.append('_')

#Fetching the age details    
try:
    for n in driver8.find_elements_by_xpath('//div[@class="age"]'):
        age8.append(n.text)
except NoSuchElementException:
    age8.append('_') 
    
#Fetching the source details    
try:
    for o in driver8.find_elements_by_xpath('//div[@class="source"]'):
        source8.append(o.text)
except NoSuchElementException:
    source8.append('_')    

#Fetching the industry details    
try:
    for s in driver8.find_elements_by_xpath('//div[@class="category"]'):
        industry8.append(s.text)
except NoSuchElementException:
    industry8.append('_')    
    
time.sleep(3)

#creating the dataframe
forbes8=pd.DataFrame({})
forbes8['Name']=Name8
forbes8['World Rank']=Rank8
forbes8['Total Net Worth']=networth8
forbes8['Country Citizon']=citizion8 
forbes8['Age']=age8
forbes8['Source of income']=source8
forbes8['Industry']=industry8

#printing the dataframe
forbes8
```

#9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
#importing the driver from the crome
driver9=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the webpage url:
url=('https://www.youtube.com/watch?v=Xjv1sY630Uc')
driver9.get(url)
time.sleep(10)


#scrolling the page to find 500 comments
for i in range(50):
    driver9.execute_script(f"window.scrollTo(0, {2**127});") 
    time.sleep(10)
    
time.sleep(10)    

#fetching the value from the page
comment9=[]
upvote9=[]
commentposted9=[]

#fetching the comments
for i in driver9.find_elements_by_id('content-text'):
    comment9.append(i.text.replace('\n',''))

#fetching the upote   
for k in driver9.find_elements_by_id('vote-count-middle'):
    upvote9.append(k.text) 

#fetching the time when commnet posted  
for m in driver9.find_elements_by_xpath('//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]//a'):
    commentposted9.append(m.text) 
    
    
time.sleep(3)
#creating the dataframe
youtube9=pd.DataFrame({})
youtube9['Comments']=comment9[:500]
youtube9['Upvote']=upvote9[:500]
youtube9['Comment posted on']=commentposted9[:500]   

#Printing Dataframe
youtube9
```

#10 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
#importing the driver from the crome
driver10=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the webpage url:
url=('https://www.hostelworld.com')
driver10.get(url)

time.sleep(3)

#entering the input element in the search box
flip_sear=driver10.find_element_by_xpath('//input[@class="location-text"]')
flip_sear.send_keys('London')
time.sleep(3)
flip_sear.clear()
flip_sear.send_keys('London')

time.sleep(5)
London10=driver10.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]')
London10.click()
time.sleep(3)

#Click on search
btn10=driver10.find_element_by_id('search-button')
btn10.click()

time.sleep(7)

#Declaring variable
hotalname10=[]
distnce10=[]
reviewcount10=[]
pageurl10=[]
price10=[]  
    
while(True):

    # Prices    
    prices = driver10.find_elements_by_xpath('//div[@class="price-col"]')
    for p in prices:
        price10.append(p.text.replace('\n',' '))
      
    
     # Hotel name
    for i in driver10.find_elements_by_xpath('//h2[@class="title title-6"]'):
        hotalname10.append(i.text)
       

    #Distance
    for j in driver10.find_elements_by_xpath('//span[@class="description"]'):
        distnce10.append(j.text)  
    
    
    #Reivew
    for l in driver10.find_elements_by_xpath('//div[@class="reviews"]'):
        reviewcount10.append(l.text)   
    
    
    #feting page url    
    for j in driver10.find_elements_by_xpath('//h2[@class="title title-6"]//a'):
        pageurl10.append(j.get_attribute('href')) 
    time.sleep(2)
    
    try: 
        next_button = driver10.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
        
# Separate  Privates_From price  and  Dorms_From price
Privates_From = []
for i in range(0,len(price10),2):
    Privates_From.append(price10[i])
time.sleep(2)

Dorms_From = []
for i in range(1,len(price10),2):
    Dorms_From.append(price10[i])
    
time.sleep(5)

#fatching data by page
rating10=[]
Totalreview10=[]
hotelfacility10=[]
propertydesc10=[]
overallriew10=[]

for a in pageurl10:
    driver10.get(a)
    time.sleep(3)   
    
    #feching the rating
    try:
        rat10=driver10.find_element_by_xpath('//div[@class="rating rating-summary-container big clickable"]//div')
        rating10.append(rat10.text.replace('\n',''))
    except NoSuchElementException:
        rating10.append('_')  
    
    #feching the facility details
    try:
        facility10=driver10.find_element_by_xpath('//div[@class="rating-factors rating-factors big"]')
        hotelfacility10.append(facility10.text.replace('\n',''))
    except NoSuchElementException:
        hotelfacility10.append('_') 
    
     #feching the property desc details
    try:
        Desc10=driver10.find_element_by_xpath('//div[@class="content collapse-content"]')
        propertydesc10.append(Desc10.text.replace('\n',''))
    except NoSuchElementException:
        propertydesc10.append('_')
    
     #feching the overall rating details
    try:
        overall10=driver10.find_element_by_xpath('//div[@class="keyword"]//span')
        overallriew10.append(overall10.text)
    except NoSuchElementException:
        overallriew10.append('_')

time.sleep(3)        
#creating the dataframe
df10=pd.DataFrame({})
df10['Hotel Name']=hotalname10
df10['Distance from city center']=distnce10
df10['Rating']=rating10
df10['Overall Review']=overallriew10
df10['Total Review']=reviewcount10 
df10['Totel Facility']=hotelfacility10
df10['Private Price']=Privates_From
df10['Drom Price']=Dorms_From
df10['Discription']=propertydesc10

#printing the dataframe
df10    
```
