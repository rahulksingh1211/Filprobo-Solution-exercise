```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. 
#You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#importing the driver for Chrome
driver1=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the webpage using url 
url=('https://www.naukri.com/')
driver1.get(url)

time.sleep(3)

#first getting the code for to Enter “Data Analyst” in “Skill,Designations,Companies” field
search_job1=driver1.find_element_by_xpath('//input[@class="sugInp"]')
search_job1.send_keys('Data Analyst')

#searching the job loc
Loc_search1=driver1.find_element_by_id('qsb-location-sugg')
Loc_search1.send_keys('Bangalore')

time.sleep(3)

#Clicking on the search button.
btn1=driver1.find_element_by_class_name('btn')
btn1.click()

time.sleep(3)

#fetching the data
job_name1=[]
job_exp1=[]
job_loc1=[]
Company_name1=[]
#fetching job name data
try:
    for i in driver1.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
        job_name1.append(i.text)
except NoSuchElementException:
    job_name1.append('NA')

#Fetching Experience data:
try:
    for j in driver1.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]//span'):
        job_exp1.append(j.text)
except NoSuchElementException:
    job_exp1.append('NA')
    
    
#Fetching Location:
try:
    for k in driver1.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]//span'):
        job_loc1.append(k.text)
except NoSuchElementException:
    job_loc1.append('NA')
    

#Fetching Location:
try:
    for m in driver1.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]'):
        Company_name1.append(m.text)
except NoSuchElementException:
    Company_name1.append('NA')

time.sleep(3)    

#creating the dataframe
Naukari_1=pd.DataFrame({})
Naukari_1['Job Title']=job_name1[:10]
Naukari_1['Company Name']=Company_name1[:10]
Naukari_1['Working Experience']=job_exp1[:10]
Naukari_1['JOb Location']=job_loc1[:10]

#making data at the center
Naukari_1=Naukari_1.style.set_properties(**{'text-align': 'center'}).set_table_styles([ dict(selector='th', props=[('text-align', 'center')])])
Naukari_1
```
