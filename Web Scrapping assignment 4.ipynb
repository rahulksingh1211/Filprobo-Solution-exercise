1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#importing the driver from the crome
driver=webdriver.Chrome(r'D:\Datatrained\Web Scrapping\chromedriver.exe')

#opening the web url:
url=('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.get(url)
time.sleep(5)

## creating empty lists for scraping data
video_rank1=[]
videoName1=[]
Uploader1=[]
video_views1=[]
uploaddate1=[]

time.sleep(4)
#scraping the video-Rank
rank1=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank1:
    if i.text is None :
        video_rank1.append("_") 
    else:
        video_rank1.append(i.text.replace('.',''))

time.sleep(4)
#scraping the video-Name
videon1=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in videon1:
    if i.text is None:
        videoName1.append("_")
    else:
        videoName1.append(i.text)

time.sleep(4)
#scraping the Uploader details
uplodr1=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in uplodr1:
    if i.text is None:
        Uploader1.append("_")
    else:
        Uploader1.append(i.text)  

time.sleep(4)
#scraping the total views
views1=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in views1:
    if i.text is None:
        video_views1.append("_")
    else:
        video_views1.append(i.text)  

time.sleep(4)
#scraping the Upload date
date1=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in date1:
    if i.text is None:
        uploaddate1.append("_")
    else:
        uploaddate1.append(i.text)   

time.sleep(3)
#creating the dataframe
df1=pd.DataFrame({})
df1['Rank']=video_rank1[:30]
df1['Video Name']=videoName1[:30]
df1['Artist']=Uploader1[:30]
df1['Upload Date']=uploaddate1[:30]
df1['Views(in billions)']=video_views1[:30]


#printing dataframe
df1
```
